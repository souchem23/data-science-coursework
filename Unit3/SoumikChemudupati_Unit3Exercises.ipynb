## The First Four Moments: Mean, Variance, Skew, and Kurtosis

When we talk about a probablity dist, we often care about at four things:
- The expected value, or where a distribution is centered (mean)
- The amount of variation from that centered value (variation)
- Whether or not our distribution favors, leans towards, or skews towards certain values. (skew)
- How often to expect unusually high or low values. (kurtosis)

**Task10**:

Describe what processes "generate" each of the previous discrete distributions models. Also report their support.

Here's a spicier way to phrase the question: describe these distributions' maximum entropy conditions in english.

Hint: Make sure to read any of the: preliZ documentation, relevant wikipedia articles, or wolfram alpha.


Bernoulli: This distribution supports x value of 1 and 0, where x=1 represents the probability of a certain event occuring and x=0 represents the probability of an event NOT occuring. This is based on a "p" argument which encodes the probability value that is used for this distribution.

Binomial: This distribution supports x values form 0 to a value "n" that is provided in the declaration of the distribution. This distribution is encoding the number of times that the two possibilities in the scenario will occur given a probability ("p") provided in the distribution declaration.

Poisson: The Possion distribution supports natural numbers (greater than 0) as part of ts range. Thi distribution, specifically, models the number of times an event will occur in a certain time frame specified. The "mu" paramter in the distribution declaration identifies the number of times an event will occur in a given interval.

Geometric: The Geometric distribution supports all natural numbrs greater than zero. Given x number of sequences, this distribution maps the probability that an event will be "successfull" out of those Bernoulli trails. The "P" encodes the probability and is declared in the class declaration.

HyperGeometric: This distribution is quite similar to the previous Geometric distribution in that it describes the probaility of a successful event occuring out of a certain number of trails. It differs in the fact that it has additional paramters such as the number of trials ("n" bernoulli trials), the population of possibilities without replacement ("N" objects), and the number of good outcomes in the population ("k"). Theis model supporsts x values within the range [max(0, n-N+k), min(k,n)].
