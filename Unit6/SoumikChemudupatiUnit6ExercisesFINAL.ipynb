**Task8**:

Describe the differences and similarities between the following three models: linear regression with two predictors, one of which is a categorical variable:

- adding the variables in the model, as is standard.
- using that categorical variable as a hierarchy upon the other predictor variable.
- adding the variables, plus the categorical variable's interaction with the other variable.

For these different models, it is important to understand how the concept of global slope plays a role.

For the linear regression with two predictors in the standard method, the model will result in different intercepts on the graph since there are different categories. However, since we are not explicitly defining a categorical variable, there will be a common global slope for all of the variables in the model.

For the linear regression with two predictors *using the categorical variable as a heirarchy*, each of the categories/variables will have different slopes and different intercepts. Heirarchical models have some more flexibility because the models can have differents slopes within a certain range. It can be in this range of the global slope and does not have to be exact.

For the linear regression with two predictors *adding the categorical interaction*, no more concept of global slope really exists. So, we need to be careful. Each slope is affected by the slop of the other variable, but there is NO global slope that must be met for the model. It is not as flexible as doing a separate linear regression for each predictor, but still provides a lot of flexibility.

````python
import numpy as np
import pandas as pd
import arviz as az
import pymc as pm
import matplotlib.pyplot as plt
import bambi as bmb
````

````python
#have to drop incomplete rows, so that bambi will run
bb = pd.read_csv(
    'https://raw.githubusercontent.com/thedarredondo/data-science-fundamentals/refs/heads/main/Data/basketball2324.csv').dropna()
````

````python
#only look at players who played more than 600 minutes
#which is 20 min per game, for 30 games
bb = bb.query('MP > 600')
#remove players who never missed a free throw
bb = bb.query('`FT%` != 1.0')
#filter out the combo positions. This will make it easier to read the graphs
bb = bb.query("Pos in ['C','PF','SF','SG','PG']")
#gets rid of the annoying '%' sign
bb.rename(columns={"FT%":"FTp","FG%":"FGp"}, inplace=True)
````

1. predictors free throw percentage and position (with position as a categorical predictor)

````python
#define the model
model_player_1 = bmb.Model("`FGp` ~ `FTp` + Pos", data=bb)
#fit the model
idata_model = model_player_1.fit(idata_kwargs={'log_likelihood': True})
````

````python
bmb.interpret.plot_predictions(model_player_1,
                               idata_model,
                                ["FTp",  "Pos"], fig_kwargs={"figsize":(11, 4)})
````

2. predictors free throw percentage and position (with position as a hierarchy)

````python
#define the model, but as a hierarchy
model_player_2 = bmb.Model("`FGp` ~ (`FTp`|Pos)", data=bb)
#create the model
idata_model_2 = model_player_2.fit(idata_kwargs={'log_likelihood': True})
````

````python
bmb.interpret.plot_predictions(model_player_2,
                               idata_model_2,
                               ["FTp","Pos"],
                               fig_kwargs={"figsize":(11, 4)})
````

3. predictors free throw percentage and position (with position interacting with frew throw percentage)

````python
#define the model; bb_interq stands for baskeball interaction quantiative
model_player_3 = bmb.Model("`FGp` ~ `FTp` + `Pos` + `FTp`:`Pos`", data=bb)
#create the model
idata_model_3 = model_player_3.fit(idata_kwargs={'log_likelihood': True})
````

````python
bmb.interpret.plot_predictions(model_player_3,
                               idata_model_3,
                               ["FTp","Pos"],
                               fig_kwargs={"figsize":(11, 13)})#,
                               #legend=False)
````

4. predictors free throw percentage, position, 3 point attempts, and interactions between all three predictors

````python
model_player_4 = bmb.Model("`FGp` ~ `FTp` + `3PA` + `Pos` + `FTp`:`3PA` + `FTp`:`Pos` + `3PA` : `Pos`", data=bb)
#create the model
idata_model_4 = model_player_4.fit(idata_kwargs={'log_likelihood': True})
````

````python
bmb.interpret.plot_predictions(model_player_4,
                               idata_model_4,
                               ["FTp","3PA","Pos"],
                               fig_kwargs={"figsize":(11, 13)})#,
                               #legend=False)
````

5. predictors free throw percentage, position, 3 point attempts, with an interaction between 3 point attempts and postion.

````python
model_player_5 = bmb.Model("`FGp` ~ `FTp` + `3PA` + `Pos` + `3PA` : `Pos`", data=bb)
#create the model
idata_model_5 = model_player_5.fit(idata_kwargs={'log_likelihood': True})
````

````python
bmb.interpret.plot_predictions(model_player_5,
                               idata_model_5,
                               ["FTp","3PA","Pos"],
                               fig_kwargs={"figsize":(11, 13)})#,
                               #legend=False)
````
